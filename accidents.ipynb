```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("Accident_train.csv")
df
```

```python
for i in range(17):
    print("Null values in Column",i,":")
    nullCheck=df.iloc[:,i].isnull().sum()
    print(nullCheck)
```

```python
df=df.dropna()
df
```

```python
for i in range(17):
    print("Null values in Column",i,":")
    nullCheck=df.iloc[:,i].isnull().sum()
    print(nullCheck)
```

```python
df=df.reset_index(drop=True)
df
```

```python
df.Special_Conditions_at_Site.unique()
```

```python
dummies1=pd.get_dummies(df['Policing_Area'])
dummies1
```

```python
dummies2=pd.get_dummies(df['Weekday_of_Collision'])
dummies2
```

```python
df.drop(['Policing_Area','Weekday_of_Collision'],axis='columns',inplace=True)
df=pd.concat([df,dummies1,dummies2],axis='columns')
df
```

```python
X=df.drop(['Collision_Severity','Collision_Ref_No'],axis='columns')
Y=df['Collision_Severity']
X
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=6)
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn1=KNeighborsClassifier(n_neighbors=3)
knn1.fit(X_train,Y_train)
Y_predicted=knn1.predict(X_test)
print("First Five Test Values:")
print(np.array(Y_test[0:5]))
print("First Five Predicted Values:")
print(Y_predicted[0:5])
print(knn1.score(X_test,Y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
knn2=KNeighborsClassifier(n_neighbors=3)
knn2.fit(X,Y)
print(cross_val_score(knn2,X,Y,cv=10))
print(np.average(cross_val_score(knn2,X,Y,cv=10)))
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3)
Y_predicted=kmeans.fit_predict(X)
Y_predicted
```

```python
df['Severity_Result']=Y_predicted
df
```
